Getting Store Names, Numbers and Addresses from Sprouts.com

```python
import requests
from bs4 import BeautifulSoup
import re  

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

states_urls = [
    "https://www.sprouts.com/stores/al/",
    "https://www.sprouts.com/stores/az/",
    "https://www.sprouts.com/stores/ca/",
    "https://www.sprouts.com/stores/co/",
    "https://www.sprouts.com/stores/de/",
    "https://www.sprouts.com/stores/fl/",
    "https://www.sprouts.com/stores/ga/",
    "https://www.sprouts.com/stores/ks/",
    "https://www.sprouts.com/stores/la/",
    "https://www.sprouts.com/stores/md/",
    "https://www.sprouts.com/stores/mo/",
    "https://www.sprouts.com/stores/nv/",
    "https://www.sprouts.com/stores/nj/",
    "https://www.sprouts.com/stores/nm/",
    "https://www.sprouts.com/stores/nc/",
    "https://www.sprouts.com/stores/ok/",
    "https://www.sprouts.com/stores/pa/",
    "https://www.sprouts.com/stores/sc/",
    "https://www.sprouts.com/stores/tn/",
    "https://www.sprouts.com/stores/tx/",
    "https://www.sprouts.com/stores/ut/",
    "https://www.sprouts.com/stores/va/",
    "https://www.sprouts.com/stores/wa/"
]

for url in states_urls:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    stores = soup.find_all('div', class_='cell medium-4 large-3 list-by-state')
    
    for store in stores:
        store_number = store.find('p', class_='store-num').text.strip()
        store_name = store.find('h4').text.strip()
        location_and_phone = ' '.join([p.text for p in store.find_all('p')[1:]]).replace('\n', ' ').strip()
        
        phone_number = re.findall(r'\d{3}-\d{3}-\d{4}', location_and_phone)
        if phone_number:
            store_location = location_and_phone.split(phone_number[0])[0].strip()  
            print(f'{store_number}; {store_name}; {store_location}; {phone_number[0]}')
        else:
            print(f'{store_number}; {store_name}; {location_and_phone}; No phone number found')
```
